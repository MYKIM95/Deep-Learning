```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import numpy as np
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
```

```python
physical_devices = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
Data_set = np.loadtxt("data/ThoraricSurgery.csv", delimiter = ",")
```

```python
Data_set
```

```python
# 총 18개의 칼럼
X = Data_set[:,:17]
y = Data_set[:,17]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 60)
```

```python
tf.random.set_seed(3)
model = Sequential()  # 신경망 생성 (한층 한층 쌓는 기능)

# add() 신경망에 layer를 추가하는 함수
model.add(Dense(10, input_dim = 17))  # Dense는 layer 를 정의하는 함수, 맨처음 dense가 선언될때만 input_dim 을 입력해주면 됨
model.add(Activation("relu"))
model.add(Dense(5))
model.add(Activation("relu"))
model.add(Dense(1))
model.add(Activation("sigmoid"))

model.summary()  # 신경망의 상태를 알려주는 함수
```

```python
model.compile(loss = "mse", optimizer = "adam", metrics  = ["acc"])
```

```python
model.fit(X_train, y_train, epochs = 300, batch_size = 5)
```

```python
w = model.get_weights()  #  이 모델이 결정한 가중치값
```

```python
w[0].shape  # 17 X 10 의 갯수를 갖는 입력층과 은닉층을 연결하는 가중치값
```

```python
w[1].shape
```

```python
w[2].shape
```

```python
model.evaluate(X_test, y_test)
```

### 1. Wine 분류하기

```python
import pandas as pd
```

```python
wine = pd.read_csv('data/wine.csv')
```

```python
wine.head()
```

```python
wine['quality'].unique()
```

### 2.1 문제와 답 나누기

```python
X= wine.drop('quality', axis = 1)
y = wine['quality']
y_en = pd.get_dummies(y)  # 정답컬럼을 7개로 원핫 인코딩
```

### 2.2 train 데이터와 test 데이터 나누기

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_en, test_size = 0.3, random_state = 60)
```

### 2.3 신경망 설계

```python
tf.random.set_seed(3)
model2 = Sequential()  # 신경망 생성 (한층 한층 쌓는 기능)

# add() 신경망에 layer를 추가하는 함수
model2.add(Dense(30, input_dim = 12))  # Dense는 layer 를 정의하는 함수, 맨처음 dense가 선언될때만 input_dim 을 입력해주면 됨
model2.add(Activation("relu"))
model2.add(Dense(15))
model2.add(Activation("relu"))
model2.add(Dense(7))
model2.add(Activation("softmax"))  # softmax 는 다중분류시 쓰는 활성화 함수 

model2.summary()  # 신경망의 상태를 알려주는 함수
```

```python
model2.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics  = ["acc"])
```

```python
model2.fit(X_train, y_train, epochs = 500, batch_size = 200)
```

### 2.4 평가

```python
model2.evaluate(X_test, y_test)
```

### 5. 보스턴 집값 예측

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(X_train, y_train ), (X_test, y_test) = boston_housing.load_data()
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train.shape
```

### 5.2 신경망설계

```python
tf.random.set_seed(3)
model5 = Sequential()  # 신경망 생성 (한층 한층 쌓는 기능)

# add() 신경망에 layer를 추가하는 함수
model5.add(Dense(500, input_dim = 13))  # Dense는 layer 를 정의하는 함수, 맨처음 dense가 선언될때만 input_dim 을 입력해주면 됨
model5.add(Activation("relu"))
model5.add(Dense(300))
model5.add(Activation("relu"))
model5.add(Dense(50))
model5.add(Activation("relu"))
model5.add(Dense(1))

model5.summary()  # 신경망의 상태를 알려주는 함수
```

```python
model5.compile(loss = "mean_squared_error", optimizer = "adam", metrics  = ["mae"])
# mse -> 평균 제곱 오차  ,  mae -> 평균 절대값 오차
```

```python
model5.fit(X_train, y_train, epochs = 1000, batch_size = 10)
```

```python
model5.predict(X_test)
```

```python
y_test
```

```python
model5.evaluate(X_test, y_test)
```

```python
y_pre = model5.predict(X_test)

for i in range(10):
    print("실제가격 : {}, 예상가격 : {}".format(y_test[i], y_pre[i][0]))
```

```python
price_diff = []

for i in range(len(y_test)):
    price_diff.append(np.abs(y_test[i] - y_pre[i][0]))
    
np.average(price_diff)
```
